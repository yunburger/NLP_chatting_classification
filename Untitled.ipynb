```python
import re
f = open("data.txt", 'r', encoding='UTF-8')
xw = open("x_train.txt", 'w', encoding = 'UTF-8')
yw = open("y_train.txt", 'w')
while True:
    line = f.readline()
    if not line: break
    #if  line[0] == '[':
    #    w.write(line)
    #    print('oo')
    if  line[0] == '[':
        temp = re.split('] ', line)
        print(temp[2])
        print(temp[0])
        if temp[0] == '[박윤보':
            yw.write('1\n')
        else :
            yw.write('0\n')
            
        xw.write(temp[2])
            
f.close()
xw.close()
yw.close()
```
